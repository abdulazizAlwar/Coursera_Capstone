```python
import pandas as pd 
import numpy as np
```

## Import Dataset

```python
df = pd.read_csv('foursquare_dubai_venues.csv')
```

```python
df.head()
```

# Get Number of each venue category as features for each adress

### One Hot Encode Categories

```python
dummies = pd.get_dummies(df['categories'])
df = pd.concat([df, dummies], axis=1).drop('categories', axis=1)
```

```python
df.head()
```

```python
df.columns
```

```python
df = df.groupby('address').sum().drop(['venue.location.lat', 'venue.location.lng'], axis=1)
```

```python
df.head()
```

```python
df.sum(axis=1).sort_values()
```

##### As seen above, the foursquare dataset does not have enough data points for proper modelling

## Train-Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test = train_test_split(df, train_size=0.7)
```

## Clustering

```python
from sklearn.cluster import KMeans
```

```python
kmeans1 = KMeans(n_clusters=3)
```

```python
kmeans1.fit(X_train)
```

## Validation

### Clustering Output on Training Dataset

```python
train_clusters = kmeans1.predict(X_train)
```

```python
X_train['predicted_clusters'] = train_clusters
```

```python
X_train.head(2)
```

### Clustering Output on Testing Dataset

```python
test_clusters = kmeans1.predict(X_test)
```

```python
X_test['predicted_clusters'] = test_clusters
```
